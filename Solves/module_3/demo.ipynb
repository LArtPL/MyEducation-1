```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set()
```

```python
def sample(locs, scales, N, n):
    samples = []
    for i in range(N):
        idx = np.random.choice(np.arange(len(locs)), size=n)
        aux = []
        for j in idx:
            aux.append(np.random.normal(locs[j], scales[j], size=1)[0])
        samples.append(np.mean(aux))
    return np.array(samples)
```

```python
locs = [0., 5., 20.]
scales = [1., 2., 3.]
n = 100 # размер выборок
N = 50000 # количество реализаций
```

### Исходное распределение

```python
S_gen = sample(locs, scales, N, 1)
sns.distplot(S_gen, );
```

### Распределение выборочных средних

```python
S = sample(locs, scales, N, n)
sns.distplot(S, );
```

### Найдем среднее и стандартное отклонение 

```python
# Среднее
```

```python
S_gen.mean()
```

```python
S.mean()
```

### Стандартное отклонение

```python
S_gen.std() / np.sqrt(n) # по формуле ЦПТ
```

```python
S.std() # Наблюдаемое стандартное отклонение выборочных средних
```
