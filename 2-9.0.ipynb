```python
students=np.array([[1,135,34,4],
                   [2,160,43,5],
                   [3,163,40,4.3],
                   [4,147,44,5],
                   [5,138,41,4.7],
                   [6,149,54,3.9],
                   [7,136,39,4.2],
                   [8,154,48,4.9],
                   [9,137,35,3.7],
                   [10,165,60,4.6]])
```

```python
mean = np.mean(students[:,-1])
mean
```

```python
median = np.median(students[:,-1])
median
```

```python
sort = np.sort(students[:,-1])
(sort[4]+sort[5])/2
```

```python
np.median(students[:,2])
```

```python
np.mean(students[:,2])-np.median(students[:,2])
```

```python
corr = np.corrcoef(students[:,1], students[:,2])
corr
```

```python
np.corrcoef(students[:,1], students[:,3]) #height and marks
```

```python
np.corrcoef(students[:,2], students[:,3]) #weight and marks
```

```python
std = np.std(students[:,1]) #std of heights
std
```

```python
np.std(students[:,3])
```

```python
np.std(students[:,2])**2  # std**2=dispersion
```
