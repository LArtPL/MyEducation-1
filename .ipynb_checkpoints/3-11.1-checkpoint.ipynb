```python
# Create dataframe for work:
log = pd.read_csv("module3_log.csv", header=None)   
log.columns = ['user_id', 'time', 'bet', 'win']  
log['time'] = log['time'][log['time'].notna()].apply(lambda x: x[1:])  
log['time'] = pd.to_datetime(log['time'])  
log['bet'] = log['bet'].fillna(0)
log['win'] = log['win'].fillna(0)
log['net'] = log['win'] - log['bet']
users['user_id'] = users['user_id'].apply(lambda x: x.lower())  
log = log[log.user_id != '#error']  
log['user_id'] = log['user_id'].str.split(' - ').apply(lambda x: x[1])  
df = pd.merge(log, users, on='user_id')  

# make data more comfortable:
df2 = df[['user_id', 'time', 'bet']] 

# Delta between min time without bet and min time with bet:
df2_time = df2[df2['bet']>0].groupby('user_id').min()-df2[df2['bet']==0].groupby('user_id').min()

df2_time['time'].mean() 
# answer: Timedelta('46 days 06:54:48')
```

```python
# Create dataframe for work:
log = pd.read_csv("module3_log.csv", header=None)   
log.columns = ['user_id', 'time', 'bet', 'win']  
log['time'] = log['time'][log['time'].notna()].apply(lambda x: x[1:])  
log['time'] = pd.to_datetime(log['time'])  
log['bet'] = log['bet'].fillna(0)
log['win'] = log['win'].fillna(0)
log['net'] = log['win'] - log['bet']
users['user_id'] = users['user_id'].apply(lambda x: x.lower())  
log = log[log.user_id != '#error']  
log['user_id'] = log['user_id'].str.split(' - ').apply(lambda x: x[1])  
df = pd.merge(log, users, on='user_id')  

# make data more comfortable:
df2 = df[['user_id', 'time', 'bet']] 

# Delta between min time without bet and min time with bet:
df2_time = df2[df2['bet']>0].groupby('user_id').min()-df2[df2['bet']==0].groupby('user_id').min()

df2_time[df2_time['time'].dt.days>0]['time'].sum()/100 # answer: Timedelta('47 days 01:20:59.460000')
```

```python
import pandas as pd
log = pd.read_csv("log.csv", header=None)   
log.columns = ['user_id', 'time', 'bet', 'win']  
users = pd.read_csv('users.csv', encoding='KOI8-R', sep='\t')
users.columns = ['user_id', 'email', 'geo']
users['user_id'] = users['user_id'].apply(lambda x: x.lower())  
log = log[log.user_id != '#error']  
log['user_id'] = log['user_id'].str.split(' - ').apply(lambda x: x[1])  
sample2 = pd.merge(log, users, on='user_id').groupby('geo')['user_id'].count()
```

```python
#1. загрузите датасеты log.csv и users.csv;
log = pd.read_csv('module3_log.csv', header=None)
log.columns = ['user_id', 'time', 'bet', 'win']
log.time = log.time[log.time.notna()].apply(lambda x: x.replace('[', ''))
us=pd.read_csv('users.csv', sep='\t', encoding='koi8_r')
us.columns=['user_id','email','geo']
# 2. удалите user_id с ошибкой (#error) и приведите признак user_id к одному виду в обоих датасетах;
us.user_id = us.user_id.apply(lambda x: x.lower())
log = log[log.user_id != '#error']  
log.user_id = log.user_id.str.split(' - ').apply(lambda x: x[1])
#3. слейте два датасета в один по признаку user_id;
log_big = pd.merge(log, us, on='user_id')
#4. сгруппируйте данные по правильному признаку (какому - вам нужно понять самим), 
# затем выберите user_id и воспользуйтесь функцией count() для подсчета наблюдений в каждой группе;
#5. результат (таблицу) запишите в sample2.
sample2 = log_big.groupby('user_id').email.count()
```
